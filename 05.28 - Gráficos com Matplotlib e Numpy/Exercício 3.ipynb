```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# https://github.com/open-numbers/ddf--gapminder--fasttrack/blob/master/countries_etc_datapoints/ddf--datapoints--pop--by--country--time.csv
df_pop = pd.read_csv("gapminder/ddf--datapoints--pop--by--country--time.csv")

# https://github.com/open-numbers/ddf--gapminder--fasttrack/blob/master/countries_etc_datapoints/ddf--datapoints--co2_terr--by--country--time.csv
df_CO2 = pd.read_csv("gapminder/ddf--datapoints--co2_terr--by--country--time.csv")

# https://github.com/open-numbers/ddf--gapminder--fasttrack/blob/master/countries_etc_datapoints/ddf--datapoints--lex--by--country--time.csv
df_estvida = pd.read_csv("gapminder/ddf--datapoints--lex--by--country--time.csv")

# https://github.com/open-numbers/ddf--gapminder--fasttrack/blob/master/ddf--entities--geo--country.csv
df_infopaises = pd.read_csv("gapminder/ddf--entities--geo--country.csv")
```

```python
# Isso é horrível, mas quando eu tentei um método elegante usando functools.reduce,
# meu python tentou reservar 60GB de memória e travou meu PC inteiro
df_combo = pd.merge(df_infopaises,
            pd.merge(df_pop,
            pd.merge(df_CO2, df_estvida
        )))
df_combo
```

### Selecionando somente dados para o ano de 2021

```python
df_2021 = df_combo[df_combo['time'] == 2021]
df_2021
```

### Haviam alguns valores "zerados" para expectativa de vida, então serão descartados porque ninguém liga pra Liechtenstein e... Hong Kong?
...Por que *Hong Kong* está em uma lista de **países**??

```python
df_2021[df_2021["lex"] == 0][["name", "lex"]]
```

```python
# Descartando "países" com valores zerados
df_2021 = df_2021[df_2021["lex"] != 0]
```

```python
# Checando que não há mais valores bizarros...
df_2021["lex"].sort_values(ascending=True)
```

### Achando valores possíveis de região, para poder fazer color coding no scatter chart

```python
df_2021["unhcr_region"].unique()
```

### Descobrindo quais países estão com NaN

```python
df_2021[df_2021["unhcr_region"].isna()]
# Taiwan!
```

### Fazendo um dicionário para atribuir cores a cada região

```python
region_to_color = {
    # somente Taiwan não tem uma região definida, vou ser político e colocar ele na ásia
    np.nan: "r",

    'unhcr_asia_pacific': "r",
    'unhcr_europe': "b",
    'unhcr_americas': "g",
    
    # vou ser político e agrupar todas as áfricas em uma cor só
    'unhcr_southern_africa': "y",
    'unhcr_middle_east_north_africa': "y",
    'unhcr_east_horn_africa_great_lates': "y",
    'unhcr_west_central_africa': "y",
}
```

### Montando o scatter chart

```python
data = {}
# emissão de co2 (milhões toneladas)
data['co2'] = np.array(df_2021["co2_terr"]) 
# estimativa de vida
data['lex'] = np.array(df_2021["lex"])
# região
data['color'] = np.array(df_2021["unhcr_region"].apply(lambda x: region_to_color[x]))
# população (milhões)
data['pop'] = np.array(df_2021["pop"] / 10e6)

fig, ax = plt.subplots(figsize=(6, 3), layout='constrained')
ax.scatter('co2', 'lex', c='color', s='pop', data=data)

ax.set_title("Tamanho = População; Cor = Região")

ax.set_xscale("log") # setar escala logarítmica
ax.set_xlabel("Emissões de CO2 (milhões de toneladas; escala logarítmica)")

ax.set_ylabel("Estimativa de vida")

ax.grid(True, which="both", alpha = 0.5)

plt.show()
```
